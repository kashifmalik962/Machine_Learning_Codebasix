```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv('IRIS.csv')
```

```python
data.rename(columns={'sepal_length':'SL','sepal_width':'SW','petal_length':'PL','petal_width':'PW'},inplace=True)
```

```python
setosa=data[data['species']=='Iris-setosa']
versicolor=data[data['species']=='Iris-versicolor']
virginica=data[data['species']=='Iris-virginica']
```

```python
plt.style.use('fivethirtyeight')
counts, bin_edges=np.histogram(setosa['PL'])
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


counts, bin_edges=np.histogram(versicolor['PL'])
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


counts, bin_edges=np.histogram(virginica['PL'])
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

```
