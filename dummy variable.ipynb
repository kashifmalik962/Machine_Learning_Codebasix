```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
plt.xlabel('travel')
plt.ylabel('SellPrice')
plt.scatter(df.Travel, df.SellPrice, color='red', marker='+')
```

```python
dummies = pd.get_dummies(df.CarModel)
dummies
```

```python
new_df = pd.concat([df,dummies], axis='columns')
new_df
```

```python
x = new_df.drop(['Mercedez Benz C class','SellPrice','CarModel'], axis='columns')
x
```

```python
y = new_df.SellPrice
y
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict([[67000,6,0,0]])
```
