```python
import pandas as pd
```

```python
df = pd.read_csv('homepricess.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
new_dummies = pd.concat([df, dummies], axis='columns')
new_dummies
```

```python
new_dummies.drop(['town','west windsor'],axis='columns', inplace=True)
```

```python
new_dummies
```

```python
x = new_dummies.drop('price', axis='columns')
x
```

```python
y = new_dummies.price
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict([[3400,0,0]])
```

```python
model.predict([[2800,0,1]])
```

```python
import joblib
joblib.dump(model, 'dummy')
```
