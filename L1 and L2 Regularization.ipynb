```python
import pandas as pd
```

```python
df = pd.read_csv('melb_data.csv')
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.YearBuilt.median()
```

```python
df.BuildingArea.median()
```

```python
df.BuildingArea = df.BuildingArea.fillna(df.BuildingArea.median())
```

```python
df.shape
```

```python
df.head()
```

```python
df.drop(['Address','Date','Postcode','CouncilArea','Lattitude',], axis='columns',inplace=True)
```

```python
df
```

```python
df.Type.unique()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.Regionname = le.fit_transform(df.Regionname)
df.SellerG = le.fit_transform(df.SellerG)
df.Suburb = le.fit_transform(df.Suburb)
df.Type = le.fit_transform(df.Type)
df.Method = le.fit_transform(df.Method)
```

```python
df
```

```python
X = df.drop('Price',axis='columns')
y = df.Price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
reg.score(X_test,y_test)
```

```python
reg.score(X_train,y_train)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```
