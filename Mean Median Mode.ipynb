```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('IRIS.csv')
```

```python
data.head(150)
```

```python
data.rename(columns={'sepal_length':'SL','sepal_width':'SW','petal_length':'PL','petal_width':'PW'},inplace=True)
```

```python
data
```

```python
data['species'].value_counts()
```

```python
iris_setosa=data[data['species']=='Iris-setosa']
iris_versicolor=data[data['species']=='Iris-versicolor']
iris_virginica=data[data['species']=='Iris-virginica']
```

```python
np.max(iris_setosa['PL'])
```

```python
iris_setosa.sort_values('PL')
```

```python
iris_setosa['PL'].value_counts()
```

# Mean

```python
print(np.mean(iris_setosa['SL']))
print(np.mean(iris_versicolor['SL']))
print(np.mean(iris_virginica['SL']))
```

```python
print(np.mean(iris_setosa['SW']))
print(np.mean(iris_versicolor['SW']))
print(np.mean(iris_virginica['SW']))
```

# Median

```python
print(np.median(iris_setosa['SW']))
print(np.median(iris_versicolor['SW']))
print(np.median(iris_virginica['SW']))
```

```python
print(np.median(np.append(iris_setosa['SW'],50)))
print(np.median(np.append(iris_versicolor['SW'],50)))
print(np.median(np.append(iris_virginica['SW'],50)))
```
